```python
import PyPDF2

# Open the PDF file
pdf_file = open('/content/RADAR- Robust AI-Text Detection via Adversarial Learning.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Initialize an empty list to store the text from each page
long_string = []

# Loop through all the pages and extract text
for page in pdf_reader.pages:
    long_string.append(page.extract_text())

# Close the PDF file
pdf_file.close()
```

```python
from textsum.summarize import Summarizer

model_name = "pszemraj/led-large-book-summary"
summarizer = Summarizer(
    model_name_or_path=model_name,
    token_batch_length=4096,
)

out_str = summarizer.summarize_string(long_string)

file_name = "/content/RADAR- Robust AI-Text Detection via Adversarial Learning.txt"

with open(file_name, "w") as file:
    file.write(out_str)

print(f"Summary saved to {file_name}")
```

```python
import os
import PyPDF2
from textsum.summarize import Summarizer

# Directory containing PDF files
pdf_directory = "/content/K-articles"
output_directory = "/content/K-summarized_articles"

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Model for summarization
model_name = "pszemraj/led-large-book-summary"
summarizer = Summarizer(
    model_name_or_path=model_name,
    token_batch_length=4096,
)

# Process each PDF file in the directory
for pdf_file_name in os.listdir(pdf_directory):
    if pdf_file_name.endswith(".pdf"):
        # Open the PDF file
        pdf_file_path = os.path.join(pdf_directory, pdf_file_name)
        with open(pdf_file_path, 'rb') as pdf_file:
            # Create a PDF reader object
            pdf_reader = PyPDF2.PdfReader(pdf_file)

            # Initialize an empty list to store the text from each page
            long_string = []

            # Loop through all the pages and extract text
            for page in pdf_reader.pages:
                long_string.append(page.extract_text())

        # Summarize the content
        out_str = summarizer.summarize_string(long_string)

        # Write the summary to a text file
        output_file_path = os.path.join(output_directory, os.path.splitext(pdf_file_name)[0] + ".txt")
        with open(output_file_path, "w") as file:
            file.write(out_str)

        print(f"Summary for {pdf_file_name} saved to {output_file_path}")
```

```python
import zipfile
import os

folder_path = '/content/K-summarized_articles'

zip_file_path = '/content/K-summarized_articles.zip'

with zipfile.ZipFile(zip_file_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            full_file_path = os.path.join(root, file)
            relative_path = os.path.relpath(full_file_path, folder_path)
            zipf.write(full_file_path, relative_path)

from google.colab import files
files.download(zip_file_path)
```
